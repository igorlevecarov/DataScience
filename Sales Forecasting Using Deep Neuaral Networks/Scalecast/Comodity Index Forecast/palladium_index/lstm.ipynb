```python
import pandas as pd
import numpy as np
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from scalecast.Forecaster import Forecaster

pd.options.display.float_format = '{:.2f}'.format
pd.options.display.max_rows = 1000

data = pd.read_csv('commodity_index.csv',parse_dates=['ts_month'])
data = data[['ts_month','palladium_index']]
data.rename(columns={"ts_month":"date","index": "palladium_index"}, inplace=True)

data.reset_index(drop=True, inplace=True)
data.head()
```

```python
print(data.shape)
data.tail()
```

```python
f = Forecaster(y=data['palladium_index'],current_dates=data['date'])
f.plot()
```

```python
f.plot_pacf(lags=20)
plt.show()
```

```python
f.set_test_length(36)
f.generate_future_dates(12)
f.set_estimator('lstm')
```

```python
f.manual_forecast(call_me='lstm',
                  lags=32,
                  batch_size=64,
                  epochs=200,
                  validation_split=.2,
                  shuffle=True,
                  activation='tanh',
                  optimizer='Adam',
                  learning_rate=0.001,
                  lstm_layer_sizes=(100,)*2,
                  dropout=(0,)*2,
                  plot_loss=True)
```

```python
f.plot_test_set(ci=True)
```

```python
f.export(to_excel=True,excel_name='results.xlsx')
f.plot(ci=True)
```

```python
f.export('model_summaries',determine_best_by='LevelTestSetMAPE')[
    ['ModelNickname',
     'LevelTestSetMAPE',
     'LevelTestSetRMSE',
     'LevelTestSetR2',
     'best_model']
]
```

```python
test_2021 = f.export_test_set_preds_with_cis('lstm')
test_2021 = test_2021.tail(12)
display(test_2021)

predicted_2022 = f.export_forecasts_with_cis('lstm')
display(predicted_2022)
```

```python
Total_2021 = test_2021['Actuals'].sum()
Total_2022 = predicted_2022['Forecast'].sum()

Yoy_growth = (Total_2022 - Total_2021)/ Total_2021 * 100

print(f'Total_2021:{Total_2021}')
print(f'Total_2022:{Total_2022}')
print(f'YOY (growth):{Yoy_growth}')
```
