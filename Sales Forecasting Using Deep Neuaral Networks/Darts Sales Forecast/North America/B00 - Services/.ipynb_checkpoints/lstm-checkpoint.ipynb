```python
import pandas as pd
import numpy as np
import pickle
import seaborn as sns
import matplotlib.pyplot as plt
from scalecast.Forecaster import Forecaster
```

```python
data = pd.read_csv('sales-data.csv',parse_dates=['date'])
```

```python
f = Forecaster(y=data['sales'],current_dates=data['date'])
f.plot()
```

```python
f.set_test_length(12)
f.generate_future_dates(12)
f.set_estimator('lstm')
```

```python
f.manual_forecast(call_me='lstm_best',
                  lags=24,
                  batch_size=16,
                  epochs=200,
                  validation_split=.2,
                  shuffle=True,
                  activation='tanh',
                  optimizer='Adam',
                  learning_rate=0.001,
                  lstm_layer_sizes=(100,)*15,
                  dropout=(0,)*15,
                  plot_loss=True)
```

```python
f.plot_test_set(ci=True)
```

```python
f.generate_future_dates(12) # forecast length of 5
f.add_time_trend()
f.export(to_excel=True,excel_name='results.xlsx')
f.plot(ci=True)
```
