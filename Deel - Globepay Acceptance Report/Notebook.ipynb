```python
import pandas as pd
import json

acceptance_data = pd.read_csv('datasets/Globepay Acceptance Report - Globepay Acceptance Report.csv')
display(acceptance_data)
print(acceptance_data.info())
print(acceptance_data['country'].unique())
print(acceptance_data['currency'].unique())
```

```python
chargeback_data = pd.read_csv('datasets/Globepay Chargeback Report - Globepay Chargeback Report.csv')
display(chargeback_data)
print(chargeback_data.info())
print(chargeback_data['chargeback'].unique())
```

# 1.1 Outlines the volume (in USD) of the declined payments



```python
def ConvertToUsd(row):
    rates = json.loads(row['rates'])
    return row['amount'] / rates[row['currency']]

acceptance_data['amountUS'] = acceptance_data.apply(lambda x: ConvertToUsd(x),axis=1)
display(acceptance_data.head())
display(acceptance_data.tail())

pd.set_option('display.float_format', lambda x: '%.5f' % x)

volumes = acceptance_data.groupby('state')['amountUS'].agg(['sum','count'])
display(volumes)

pd.options.display.float_format = '{:.2f}'.format
volumes.plot(kind='bar')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 6))

ax.pie(volumes['count'], labels=['ACCEPTED','DECLINED'], autopct='%.1f%%')
ax.set_title('Acceptance vs Decline Rates')
plt.tight_layout()
```

# 2.1. What is the acceptance rate over time?

```python
display(acceptance_data['date_time'].min())
display(acceptance_data['date_time'].max())

acceptance_data['date_time']= pd.to_datetime(acceptance_data['date_time'])
accepted_data = acceptance_data[acceptance_data.state=='ACCEPTED'].groupby(acceptance_data['date_time'].dt.month)['state'].count().reset_index()

declined_data = acceptance_data[acceptance_data.state=='DECLINED'].groupby(acceptance_data['date_time'].dt.month)['state'].count().reset_index()

sum_data =  accepted_data.merge(declined_data, on='date_time')
sum_data['total'] = sum_data['state_x'] + sum_data['state_y']
sum_data['acceptance_rate'] = (sum_data['state_x'] / sum_data['total']) * 100
display(sum_data)

plt.plot(sum_data.date_time,sum_data.acceptance_rate)
plt.xlabel("Month")
plt.ylabel("Acceptance Rate")
# function to show the plot
plt.show()
```

# 2.1 List the countries where the amount<br> of declined transactions went over $25M

```python
declinedTransactions = merged_data[merged_data.state=='DECLINED'].groupby(['state','country'])['amountUS'].sum().sort_values(ascending=False).reset_index()
display(declinedTransactions)

display(declinedTransactions[declinedTransactions.amountUS>25000000])
display(declinedTransactions[declinedTransactions.amountUS>25000000]['country'].unique())
```
